```python
if "brand" in df.columns:
    print("\nTop 10 Brands:")
    brand_count = df["brand"].value_counts().head(10)
    print(brand_count)

    plt.figure()
    brand_count.plot(kind="bar")
    plt.title("Top 10 Brands")
    plt.xlabel("Brand")
    plt.ylabel("Number of Models")
    plt.tight_layout()
    plt.show()
```

```python
# Price Analysis (if exists)
if "price" in df.columns:
    print("\nTop 5 Expensive Phones:")
    top5 = df.sort_values(by="price", ascending=False).head(5)
    print(top5)

    plt.figure()
    plt.hist(df["price"].dropna(), bins=20)
    plt.title("Price Distribution")
    plt.xlabel("Price")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

```

```python
# Battery Analysis (if exists)
if "battery" in df.columns:
    print("\nAverage Battery Capacity:", df["battery"].mean())

print("\nAnalysis Completed Successfully 🚀")
```
